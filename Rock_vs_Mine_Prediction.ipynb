Import Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
#loading the SONAR dataset to pandas Dataframe

sonar_data = pd.read_csv('sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

How much Example for Each Mine(M) and Rock(R)

```python
sonar_data[60].value_counts()
```

grouping the  Data based on rock and mine

```python
sonar_data.groupby(60).mean()
```

Saprateting data and Label

```python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

Trainig and Test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training --> Logistic regression

```python
model = LogisticRegression()
```

```python
# training logistic  regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluaction

Check accuracy score  on training data

```python
#accuracy on training data

X_train_pridiction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pridiction,Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

Accuracy on Test data

```python
X_test_pridiction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pridiction,Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```
